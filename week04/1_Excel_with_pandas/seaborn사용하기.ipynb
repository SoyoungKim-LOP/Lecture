```python
df = pd.read_excel('Data/sample-line.xlsx')
print(df.shape)
df.head()
```

```python
df.plot()
```

```python
plt.figure(figsize=(30,8))
sns.countplot(data=df, x='종목명')
```

```python
len(df.index)
```

```python
plt.figure(figsize=(30,8))
sns.countplot(data=df, x='종목명')
plt.xticks(rotation=len(df.index))
```

```python
df['종목명'].value_counts().sort_values(ascending=False)   # ascending=False: 아래에서 위로 읽어오기  
```

```python
df.groupby('종목명').size().sort_values(ascending=False)[:10]  # df.groupby() 함수 사용예
```

```python
df['종목명'].value_counts().sort_values(ascending=False)[:10]   # ascending=False: 아래에서 위로 읽어오기  
```

```python
팔레트 = sns.color_palette('husl')
series = df.groupby('종목명').size()
series.plot(kind='bar', figsize=(15,5), color=팔레트)
```

```python
팔레트 = sns.color_palette('husl')
series = df.groupby('종목명').size()
series.plot(kind='bar', figsize=(30,8), color=팔레트)
```

```python
with sns.color_palette('husl') as 팔레트:
    series = df.groupby('종목명').size()
    series.plot(kind='bar', figsize=(15,5), color=팔레트)
```
